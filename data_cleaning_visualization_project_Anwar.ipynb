# **Data** **Cleaning**

```python
import pandas as pd

# Specify the path to your CSV file
csv_file_path = '/content/Billionaires Statistics Dataset.csv'

# Create a Pandas DataFrame from the CSV file
df = pd.read_csv(csv_file_path)

# Display the first few rows of the DataFrame
print("First few rows of the Billionaires Statistics Dataset:")
display(df.head())
```

```python
# get the column
df.columns
```

```python
#rename columns
df.rename(columns={'finalWorth':'totalWorth'}, inplace=True)
```

```python
# let's explore the data

df.head()
```

```python
#check how many missing values
df.isnull().sum()
```

```python
#replace the missing value(s) with mean

df['age'] = df['age'].fillna(df['age'].mean())
df.isnull().sum()['age'] #شيكت انه ضبط الريبلايس للميسينق
```

```python
# Drop rows with missing values
df.dropna(inplace=True)
```

```python
# Check for duplicate rows
print(df.duplicated().sum())
```

```python
# Check for outliers
print(df.describe())
```

```python
# Remove outliers
df = df[~((df['totalWorth'] < 1200.000000) | (df['totalWorth'] > 160000.000000))]
#اعتبرت أي قيمة أقل من ١٢٠٠ متطرفة و كذلك الأعلى من ١٦٠٠ متطرفة
```

```python
# Check for inconsistencies
print(df.value_counts())
```

```python
# Save cleaned data to CSV file
df.to_csv('clean_data_Anwar.csv', index=False) #حفظت الداتا بعد التنظيف باسمي
```

# **Data** **Visualization** **Figures**

```python
import matplotlib.pyplot as plt

# Line Chart for Income
plt.figure(figsize=(10, 6))
plt.plot(new_df['personalFees'], new_df['monthlyIncome'], marker='o', linestyle='-', color='blue')
plt.xlabel('monthlyIncome')
plt.ylabel('personalFees')
plt.title('Income)')
plt.grid(True)
plt.show()
```
